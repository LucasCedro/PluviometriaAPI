```python
# Import Meteostat library and dependencies
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily,  Stations, Hourly
```

```python
# Set coordinates to the desired location
lat = -20.316524
lon = -53.680500

# Set time period
start = datetime(1900, 1, 1)
end = datetime(2099, 6, 1)
```

```python
# Pega a estação meteorológica mais próxima
stations = Stations()
stations = stations.nearby(lat, lon)
station = stations.fetch(10)
station
```

```python
# Pega a estação meteorológica mais próxima
stations = Stations()
stations = stations.nearby(lat, lon)
stations = stations.inventory('daily', True)
station = stations.fetch(1)
station
```

```python
data = Daily(station, start, end)
data = data.fetch()
```

```python
# Plot line chart including average, minimum and maximum temperature
data.plot(y=['tavg', 'tmin', 'tmax'])
plt.show()
```
